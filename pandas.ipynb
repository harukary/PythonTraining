# Pandas
* CSV読み込み
* データ加工
* 距離行列取得
* 重みづけランダム抽出

```python
import pandas as pd
```

## CSV読み込み

```python
df = pd.read_csv('data/food.csv', index_col=1)
display(df.head())
```

## データ加工

```python
# 置換
df = df.replace({'Tr': '0', '-': '0', '†': '', '\(': '', '\)': '', '\*': '0'}, regex=True)
# 型変換
df = df.astype(float)
# NaN埋め
df = df.fillna(0)
# 列ラベルで選択
df_mineral = df[['l_class','WATER', 'PROT-', 'NA', 'K', 'CA', 'MG', 'P', 'FE', 'ZN', 'CU', 'MN']]
# 列ラベルで削除
df_mineral = df_mineral.drop(['WATER', 'PROT-'],axis=1)
# 条件で抽出
df_mineral = df_mineral[(df['l_class'] != 18) & (df['l_class'] != 15)]
# printよりきれいに表示
display(df_mineral)
```

## CSV書き込み

```python
df_mineral.to_csv('data/mineral.csv', encoding='utf-8')
```

## 距離行列取得

```python
from scipy.spatial import distance_matrix
```

```python
d_mat = pd.DataFrame(distance_matrix(df_mineral.values, df_mineral.values), index=df_mineral.index, columns=df_mineral.index)
display(d_mat)
```

## 重みづけランダム出力

```python
import random
```

```python
# 距離でソート
d_list = d_mat['（えんどう類）　グリンピース　生'].sort_values()
# 重みづけランダム出力
candidates = d_list[1:11] # 0は同じ
w = [1/c if c != 0 else 100000 for c in candidates.to_list()] # 逆数で重みづけ
id = random.choices(candidates.index.to_list(), k=1, weights=w)[0]
print(id)
```
